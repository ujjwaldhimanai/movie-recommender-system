### Data transformation

```python
movies.iloc[0]['genres']
```

```python
import ast

def extract(row):
    try:
        # Convert string representation to an actual list and
        # Extract genre names 
        genre_names = [genre['name'] for genre in ast.literal_eval(row)]
        return genre_names
    except (ValueError, TypeError, KeyError):
        # Handle exceptions if the data is not in the expected format
        return []
```

```python
movies['genres'] = movies['genres'].apply(extract)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(extract)
```

```python
def extract_cast(row):
    try:
        # Convert string representation to an actual list and
        # Extract genre names 
        genre_names = [genre['name'] for index, genre in enumerate(ast.literal_eval(row), 1) if index <= 3 ]
        return genre_names
    except (ValueError, TypeError, KeyError):
        # Handle exceptions if the data is not in the expected format
        return []
            
```

```python
cast = movies.iloc[1].cast
```

```python
cast
```

```python
cast = extract_cast(cast)
```

```python
cast
```

```python
movies['cast'] = movies['cast'].apply(extract_cast)
```

```python
movies.head()
```

```python
def get_director(row):
    try:
        # Convert string representation to an actual list and
        # Extract genre names 
        director = [item['name'] for item in ast.literal_eval(row) if item.get("job") == "Director"]
        return director
    except (ValueError, TypeError, KeyError):
        # Handle exceptions if the data is not in the expected format
        return []
            
```

```python
movies['crew'] = movies['crew'].apply(get_director)
```

```python
movies.head()
```

```python
movies.overview[0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
genre = movies.genres[0]
genre
```

```python
''.join(genre).replace(' ', '').split(',')
```

```python
genre = [item.replace(" ", "") for item in genre]
```

```python
genre
```

```python
def remove_whitespace(row):
    try:
        return [item.replace(" ", "") for item in row]
    except:
        return []
```

```python
movies['genres'] = movies['genres'].apply(remove_whitespace)
movies['keywords'] = movies['keywords'].apply(remove_whitespace)
movies['cast'] = movies['cast'].apply(remove_whitespace)
movies['crew'] = movies['crew'].apply(remove_whitespace)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
movies['tags'] = movies['tags'].apply(lambda x: " ".join(x))
```

```python
movies.head()
```

```python
movies_df = movies[['movie_id','title','tags']]
movies_df
```

```python
type(movies_df.tags[0])
```

```python
movies_df['tags'] = movies_df['tags'].apply(lambda x: x.lower())
movies_df.head()
```

### Word stemming

```python
! pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    stemmed_text = [ps.stem(word) for word in text.split()]
    return " ".join(stemmed_text)
```

```python
movies_df['tags'] = movies_df['tags'].apply(stem)
```

```python
movies_df.tags[0]
```

### Model development

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(movies_df['tags']).toarray()
```

```python
vectors
```

```python
vectors.shape
```

```python
vectors[0]
```

```python
list(cv.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
similarity.shape
```

```python
movies_df[movies_df['title'] == "Batman"].index[0]
```

```python
sorted(list(enumerate(similarity[0])), reverse = True)
```

### Main function

```python
def recommend(movie):
    movie_index = movies_df[movies_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(movies_df.iloc[i[0]].title)
    
    
```

```python
recommend("Batman")
```
